```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
W=pd.read_csv('whiskey.csv')
```

```python
freq_itemset=apriori(df=W, min_support=0.1, use_colnames=True)
```

```python
freq_itemset.sample(10)
```

```python
rules=association_rules(freq_itemset, min_threshold=0.5)
```

```python
rules
```

```python
rules.sort_values(by='lift', ascending=False)
```
